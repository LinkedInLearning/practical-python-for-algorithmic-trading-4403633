## Load the data

```python
import pandas as pd

df = pd.read_excel('data/Microsoft_LinkedIn_Processed.xlsx', parse_dates=['Date'], index_col=0)
df
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define the split point
split_point = int(len(df) * 0.8)

# Split the data into training and testing sets
train_data = df.iloc[:split_point]
test_data = df.iloc[split_point:]

# Define the features and target variable
features = ['Open', 'High', 'Low', 'Close', 'Volume']
target = 'change_tomorrow'

# Create the training and testing feature matrices and target vectors
X_train = train_data[features]
y_train = train_data[target]
X_test = test_data[features]
y_test = test_data[target]

# Create a Decision Tree Regressor
dt_regressor = DecisionTreeRegressor(max_depth=15)

# Train the Decision Tree Regressor
dt_regressor.fit(X_train, y_train)

# Make predictions for each day in the testing data
predictions = []
for i in range(len(X_test)):
    # Get the features for the current day
    current_features = X_test.iloc[[i]]
    
    # Make a prediction for the current day
    current_prediction = dt_regressor.predict(current_features)[0]
    predictions.append(current_prediction)
    
    # Update the training data with the current day's data
    train_data = train_data.iloc[1:].copy()
    train_data = pd.concat([train_data,test_data.iloc[[i]]])
    X_train = train_data[features].copy()
    y_train = train_data[target].copy()
    
    # Retrain the Decision Tree Regressor on the updated training data
    dt_regressor.fit(X_train, y_train)
```

```python
df_pred = pd.DataFrame({
    'y_test': y_test,
    'y_pred': predictions,
})
```

```python
import numpy as np
```

```python
df_pred['error'] = df_pred.diff(axis=1).y_pred
```

```python
class Regression(Strategy):

    n_train = 1000
    limit_buy = 1
    limit_sell = -5
    
    def init(self):
        self.already_bought = False
        self.model = DecisionTreeRegressor(max_depth=15, random_state=42)
        
        df = self.data.df.iloc[:self.n_train]
        y = df['change_tomorrow']
        X = df.drop(columns='change_tomorrow')
        
        self.model.fit(X, y)
        
    def next(self):
        
        if len(self.data) < self.n_train:
            return
        
        X = self.data.df.drop(columns='change_tomorrow')
        X_tomorrow = X.iloc[[-1], :]
        
        forecast_tomorrow = self.model.predict(X_tomorrow)[0]
        # print(forecast_tomorrow)
        
        # conditions to sell or buy
        if forecast_tomorrow > self.limit_buy/5 and self.already_bought == False:
            self.buy()
            self.already_bought = True
        elif forecast_tomorrow < self.limit_sell/5 and self.already_bought == True:
            self.sell()
            self.already_bought = False
        else:
            pass
```

```python
class RegressionWalkForward(Regression):
    def next(self):
        if len(self.data.df) < self.n_train:
            return
        
        # train every 20 days
        if len(self.data) % 20 == 0:
            return super().next()
        
        df = self.data.df.iloc[-self.n_train:, :]
        
        X = df.drop(columns='change_tomorrow')
        y = df['change_tomorrow']

        self.model.fit(X, y)
        
        # print(f'model retrained from {df.index.min()} to {df.index.max()}')
        
        super().next()
```

The backtest is taking decisions within the class; in other words:

1. Retrains the data
2. Takes decisions for the next 20 points 
3. Retrains taking into account the past 20 points
4. And on and on and on...

```python
import backtesting
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA, GOOG

class MyStrategy(Strategy):
    def init(self):
        self.ma1 = self.I(SMA, self.data.Close, 10)
        self.ma2 = self.I(SMA, self.data.Close, 20)
        
    def next(self):
        if crossover(self.ma1, self.ma2):
            self.buy()
        elif crossover(self.ma2, self.ma1):
            self.sell()
            
# Define the time period for the backtest
start = '2010-01-01'
end = '2020-12-31'

# Define the walk forward period
period = 252 * 2 # Two years

# Define the number of periods to test
n_periods = 5

# Define the number of periods to train on
n_train_periods = 3

# Define the data to use for the backtest
data = GOOG[start:end]

# Perform the walk forward analysis
results = []
for i in range(n_periods):
    start_date = data.index[0] + pd.DateOffset(days=i * period)
    end_date = start_date + pd.DateOffset(days=period)
    
    train_start = start_date
    train_end = start_date + pd.DateOffset(days=n_train_periods * period)
    
    test_start = train_end
    test_end = end_date
    
    train_data = data.loc[train_start:train_end]
    test_data = data.loc[test_start:test_end]
    
    bt = Backtest(train_data, MyStrategy)
    bt.run()
    
    results.append(bt.run(test_data))
    
# Combine the results of each walk forward period
final_results = backtesting.combine(results)

# Print the final results
print(final_results)
```
